```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("loan_small.csv")
```

```python
data
```

```python
data2 = data.copy()
```

```python
data2 = data2.dropna(subset = ['Loan_Status'])
```

```python
data2
```

```python
cleandata = data2
```

```python
df = cleandata.copy()
```

```python
df
```

```python
cols = ['Gender' , 'Area' , 'Loan_Status']
df[cols] = df[cols].fillna(df.mode().iloc[0])
df.isnull().sum(axis = 0)
```

```python
cols2 = ['ApplicantIncome' , 'CoapplicantIncome' , 'LoanAmount']
df[cols2] = df[cols2].fillna(df.mean())
df.isnull().sum(axis = 0)
```

```python
df
```

```python
df = df.drop(['Loan_ID'],axis = 1)
```

```python
df
```

```python
df = pd.get_dummies(df, drop_first = True)
```

```python
df
```

```python
#SPlit the data vertically into X and Y
```

```python
X = df.iloc[:, :-1] # Except Loan status column
Y = df.iloc[:,-1] #only Loan Status
```

```python
#Split the data set by rows
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.2,random_state = 1234)
```
