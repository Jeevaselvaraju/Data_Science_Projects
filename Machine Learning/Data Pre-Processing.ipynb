```python
dt = df.copy()
```

```python
#replace categorical values with mode 
```

```python
cols = ['Gender','Area', 'Loan_Status']
```

```python
dt[cols] = dt[cols].fillna(dt.mode().iloc[0])
```

```python
dt[cols].shape
```

```python
dt.isnull().sum(axis = 0)
```

```python
#replace numerical valuse with mean
```

```python
cols2 = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']
```

```python
dt[cols2] = dt[cols2].fillna(dt.mean())
```

```python
dt.isnull().sum(axis = 0)
```

```python
#Label Encoding
```

```python
dt.dtypes
```

```python
#change data types
```

```python
colss = ['Gender','Area', 'Loan_Status']
```

```python
dt[colss] = dt[colss].astype('category')
```

```python
dt.dtypes
```

```python
for columns in colss:
    dt[columns] = dt[columns].cat.codes
```
