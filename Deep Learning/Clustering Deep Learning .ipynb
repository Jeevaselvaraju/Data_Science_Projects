```python
import pandas as pd
```

```python
dataset = pd.read_csv("studentclusters.csv")
```

```python
X=dataset.copy()
```

```python
#visualise the data using pandas
```

```python
X.plot.scatter(x='marks',y = 'shours')
```

```python
 ##K -means Users Euclidience distance 

#Euclidinace dist b/w2 points
```

```python
#Normalize the data using standard and min max
```

```python
from sklearn.preprocessing import minmax_scale
```

```python
X_scaled = minmax_scale(X)
```

```python
X_scaled 
```

```python
#Import KMeans
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3,random_state=1234)
```

```python
#Fit the data to get the clusters
```

```python
kmeans.fit(X_scaled)
```

```python
labels = kmeans.labels_
```

```python
labels
```

```python
#Visualise the Clusters
```

```python
labels = pd.DataFrame(labels)
```

```python
df = pd.concat([X, labels], axis =1 )
df
```

```python
df = df.rename(columns={0:'labels'})
```

```python
df
```

```python
df.plot.scatter(x = 'marks', y='shours', c='labels', colormap='Set1')
```

```python
#Elbow Method for optimum clusters
```

```python
inertia = []
for i in range(2,16):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(2,16),inertia, marker = 'o')
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel('Squared Sum(Inertia)')
```
